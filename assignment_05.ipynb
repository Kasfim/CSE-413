```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import kstest, norm

temp_data = np.array([15, 17, 16, 14, 18, 20, 19, 15, 21, 22])
print("Recorded Temperatures:", temp_data)


mean_temp = np.mean(temp_data)
std_temp = np.std(temp_data, ddof=1)
print(f"Mean Temperature = {mean_temp:.2f}, Standard Deviation = {std_temp:.2f}")

normalized_temp = (temp_data - mean_temp) / std_temp

ks_outcome = kstest(normalized_temp, 'norm')
print(f"Test Decision: {int(ks_outcome.pvalue < 0.05)}")
print(f"p-value: {ks_outcome.pvalue:.4f}")
print(f"KS Statistic: {ks_outcome.statistic:.4f}")

ordered_data = np.sort(normalized_temp)
empirical_cdf = np.linspace(1 / len(normalized_temp), 1, len(normalized_temp))

standard_cdf = norm.cdf(ordered_data)

plt.figure(figsize=(8, 5))
plt.plot(ordered_data, empirical_cdf, label='Observed CDF', color='green', linewidth=2)
plt.plot(ordered_data, standard_cdf, label='Theoretical Normal CDF', color='purple', linestyle='--', linewidth=2)
plt.title("K-S Test Comparison: Empirical vs Normal Distribution (Temperature Data)")
plt.xlabel("Normalized Temperature Values")
plt.ylabel("Cumulative Probability")
plt.legend(loc='lower right')
plt.grid(True)
plt.show()
```
