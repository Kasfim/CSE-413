```python
import numpy as np
import matplotlib.pyplot as plt

vector1 = np.random.rand(15)
vector2 = np.random.rand(15)

plt.plot(vector1, label='Vector 1')
plt.plot(vector2, label='Vector 2')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.title('Plot of Two Random Vectors')

plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

matrix = np.random.rand(4, 4)

sns.heatmap(matrix, annot=True, cmap='viridis', fmt=".2f")

plt.xlabel("Columns")
plt.ylabel("Rows")
plt.title("Heatmap of a 4x4 Matrix")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

matrix1 = np.random.rand(4, 4)
matrix2 = np.random.rand(4, 4)

addition = matrix1 + matrix2
subtraction = matrix1 - matrix2
multiplication = np.dot(matrix1, matrix2)

addition_flat = addition.flatten()
subtraction_flat = subtraction.flatten()
multiplication_flat = multiplication.flatten()

plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.bar(range(len(addition_flat)), addition_flat)
plt.title("Addition")

plt.subplot(1, 3, 2)
plt.bar(range(len(subtraction_flat)), subtraction_flat)
plt.title("Subtraction")

plt.subplot(1, 3, 3)
plt.bar(range(len(multiplication_flat)), multiplication_flat)
plt.title("Multiplication")

plt.tight_layout()
plt.show()
```
